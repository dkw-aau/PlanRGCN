```python
from graph_construction.qp.visitor.AbstractVisitor import *
from graph_construction.qp.visitor.DefaultVisitor import DefaultVisitor
import json5, json, os

class DatExtractor(DefaultVisitor):
    def __init__(self):
        self.preds = list()
        self.ents = list()
        
    def visitTriple(self,input):
        if 'http' in input['Predicate']:
            self.preds.append(input['Predicate'])
        if 'http' in input['Subject']:
            self.ents.append(input['Subject'])
        if 'http' in  input['Object']['value']:
            self.ents.append(input['Object']['value'])
    def visitPath(self,input):
        if 'Predicates' in input.keys() and 'http' in input['Predicates'][0]:
            self.preds.append(input['Predicates'][0])
        if 'http' in  input['Subject']:
            self.ents.append(input['Subject'])
        if 'http' in  input['Object']['value']:
            self.ents.append(input['Object']['value'])



def get_ent_rels(query_file):
    visitor = DatExtractor()
    try:
        qp = json.load(open(query_file, "r"))
    except Exception:
        try:
            qp = json5.load(open(query_file, "r"))
        except Exception as e:
            print(query_file)
            print(e)
    dispatch(qp,visitor)
    return visitor.preds, visitor.ents

ent_dict = {}
rel_dict = {}
for idx, row in plan_pp_df.iterrows():
    f = os.path.join('/data/wikidata_3_class_full/queryplans',idx)
    rels, ents = get_ent_rels(f)
    ent_dict[idx] = ents
    rel_dict[idx] = rels
```

```python
plan_pp_df =plan_pp_df.reset_index()
```

```python
plan_pp_fast_true = plan_pp_df[(plan_pp_df['time_cls']==0)&(plan_pp_df['planrgcn_prediction']==0)]
plan_pp_med =plan_pp_df[(plan_pp_df['time_cls']==2)]
```

```python
rel_pps_f_m = {}
rel_pps_m = {}
for idx, row in plan_pp_fast_true.iterrows():
    for r in rel_dict[row['id']]:
        if r in rel_pps_f_f.keys():
            rel_pps_f_f[r] = rel_pps_f_f[r]+ 1
        else:
            rel_pps_f_f[r] = 1
            
for idx, row in plan_pp_med.iterrows():
    for r in rel_dict[row['id']]:
        if r in rel_pps_m.keys():
            rel_pps_m[r] = rel_pps_m[r]+ 1
        else:
            rel_pps_m[r] = 1
```
