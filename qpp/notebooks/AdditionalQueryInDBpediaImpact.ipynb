```python
import pandas as pd
```

```python
old_dbpedia = '/data/DBpedia2016_0_1_10_weight_loss'
new_dbpedia = '/data/DBpedia2016_0_1_10_v3_weight_loss'
```

```python
def load_ids(p):
    ids = []
    for p_p in [f'{p}/train_sampled.tsv',f'{p}/val_sampled.tsv',f'{p}/test_sampled.tsv']:
        df = pd.read_csv(p_p, sep='\t')
        ids.extend(df['queryID'])
    return ids
old_ids = load_ids(old_dbpedia)
new_ids = load_ids(new_dbpedia)
```

```python
len(set(old_ids)), len(set(new_ids))
```

```python
diff = set([x for x in new_ids if x not in old_ids])
```
